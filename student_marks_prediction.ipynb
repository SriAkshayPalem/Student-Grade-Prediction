```python
#split dataset
#matrix = X = Independent Variable = student_hours
#vector = y = Dependent Variable = student_marks 
X=data1.drop("student_marks",axis="columns")
y=data1.drop("study_hours",axis="columns")
print("shape of X =",X.shape)
print("shape of y =",y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=51) # taking 20% data for test and another 20% is randomly
print("shape of X_train =",X_train.shape)
print("shape of y_train =",y_train.shape)
print("shape of X_test =",X_test.shape)
print("shape of y_test =",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression #linearRegression y=m*x+c
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.coef_#To know value of m 
```

```python
lr.intercept_ #To know value of c 
```

```python
m=3.93571802
c=50.44735504
y=m*3+c #Given x value as 3
y
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
pd.DataFrame(np.c_[X_test,y_test,y_pred],columns=["study_hours","student_marks_original Percentage","student_marks_predicted Percentage"])
    #Converting this into dataframe to do data analysis
```
